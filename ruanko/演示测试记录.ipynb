## 第 1 步：查看系统数据库

**功能**：显示系统自带的 Ruanko 系统数据库


```python
SHOW DATABASES;
```

## 第 2 步：创建用户数据库

**功能**：创建名为 shop 的用户数据库


```python
CREATE DATABASE shop;
```

## 第 3 步：切换数据库

**功能**：切换到 shop 数据库


```python
USE shop;
```

## 第 4 步：创建表（覆盖 5 种数据类型）

**功能**：创建 products 表，覆盖 INTEGER/VARCHAR/DOUBLE/DATETIME/BOOL


```python
CREATE TABLE products (
    id INTEGER PRIMARY KEY,
    name VARCHAR(100) NOT NULL,
    price DOUBLE,
    stock INTEGER DEFAULT 0,
    created_at DATETIME,
    is_active BOOL
);
```

## 第 5 步：查看当前数据库的表

**功能**：显示 products 表（字段数 6，记录数 0）


```python
SHOW TABLES;
```

## 第 6 步：插入数据

**功能**：批量插入 4 条商品记录


```python
INSERT INTO products VALUES
(1, 'iPhone 15', 5999.00, 100, '2024-01-15 10:30:00', true),
(2, 'MacBook Pro', 14999.00, 50, '2024-02-20 14:00:00', true),
(3, 'AirPods Pro', 1999.00, 200, '2024-03-10 09:00:00', true),
(4, 'iPad Mini', 3999.00, 0, '2024-01-20 11:00:00', false);
```

## 第 7 步：全表查询

**功能**：SELECT * FROM products


```python
SELECT * FROM products;
```

## 第 8 步：条件查询 - 比较运算

**功能**：WHERE price > 5000 与 WHERE stock = 0


```python
SELECT name, price FROM products WHERE price > 5000;
```

```python
SELECT * FROM products WHERE stock = 0;
```

## 第 9 步：条件查询 - 逻辑运算 AND

**功能**：WHERE stock > 0 AND price < 10000


```python
SELECT name, price, stock FROM products WHERE stock > 0 AND price < 10000;
```

## 第 10 步：条件查询 - 逻辑运算 OR

**功能**：WHERE stock = 0 OR is_active = false


```python
SELECT * FROM products WHERE stock = 0 OR is_active = false;
```

## 第 11 步：更新记录

**功能**：UPDATE 修改库存/价格，并验证


```python
UPDATE products SET stock = 150 WHERE id = 1;
```

```python
UPDATE products SET price = 13999.00, stock = 80 WHERE id = 2;
```

```python
SELECT * FROM products WHERE id = 1 OR id = 2;
```

## 第 12 步：删除记录

**功能**：DELETE 指定 id，并验证


```python
DELETE FROM products WHERE id = 4;
```

```python
SELECT * FROM products;
```

**修复记录**：
1. `pkg/parser/parser.go` 的 `parseCreateUser` 原先按 `TOKEN_IDENTIFIER` 匹配 PASSWORD，但 PASSWORD 已注册为关键字 `TOKEN_PASSWORD`，修正后 `CREATE USER ... PASSWORD '...'` 可正常解析。
2. `parseRestore` 原先强制要求 `TO`，修正后同时兼容 `FROM`/`TO`，使 `RESTORE DATABASE ... FROM ...` 可正常解析。

**已知简化项**：
- DATETIME 字段显示为 `0049-11-30 00:00:00`，可能为日期存储/解析未完全实现。
- ROLLBACK 为简化实现，不会撤销事务中的插入。
